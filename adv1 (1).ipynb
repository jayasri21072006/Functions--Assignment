
---

**1. What is a random variable in probability theory?**
A random variable assigns numerical values to outcomes of a random process.
It can be discrete (countable) or continuous (uncountable).
It helps quantify randomness in experiments.
Examples: number of heads in a coin toss, time to run a mile.
It's fundamental in defining probability distributions.

---

**2. What are the types of random variables?**
There are two main types: discrete and continuous.
Discrete variables take specific values (e.g., 0, 1, 2...).
Continuous variables take values in an interval (e.g., any real number between 0 and 1).
Each type uses different probability models.
They define how probabilities are assigned.

---

**3. What is the difference between discrete and continuous distributions?**
Discrete distributions assign probabilities to exact values.
Continuous distributions assign probabilities over intervals.
In discrete, P(X = x) > 0; in continuous, P(X = x) = 0.
PDFs apply to continuous, PMFs to discrete.
Examples: Binomial (discrete), Normal (continuous).

---

**4. What are probability distribution functions (PDF)?**
PDF gives the likelihood of a random variable taking a range of values.
It’s used for continuous variables.
The area under the PDF curve represents probability.
It must be non-negative and integrate to 1.
It models data like heights, weights, etc.

---

**5. How do cumulative distribution functions (CDF) differ from PDFs?**
CDF gives the probability that a variable is ≤ x.
PDF gives the density at a specific point (for continuous).
CDF is the integral of the PDF.
CDF always increases and ranges from 0 to 1.
CDF works for both discrete and continuous variables.

---

**6. What is a discrete uniform distribution?**
All outcomes are equally likely.
Used when each of 'n' outcomes has the same probability (1/n).
Example: Rolling a fair die.
Mean = (a + b)/2, where a and b are bounds.
It is a basic discrete distribution.

---

**7. What are the key properties of a Bernoulli distribution?**
It models a single trial with two outcomes: success (1) or failure (0).
Probability of success is *p*, failure is *1-p*.
Mean = p; Variance = p(1-p).
Used in yes/no type experiments.
Basis for the binomial distribution.

---

**8. What is the binomial distribution, and how is it used in probability?**
It models the number of successes in *n* independent Bernoulli trials.
Requires fixed *n*, probability *p*, and binary outcomes.
P(X = k) = C(n, k)·p^k·(1-p)^(n-k).
Used in quality control, survey analysis.
Mean = np; Variance = np(1-p).

---

**9. What is the Poisson distribution and where is it applied?**
It models the number of events in a fixed interval of time or space.
Events must be rare, independent, and occur at a constant average rate (λ).
Used in call centers, traffic flow, etc.
Mean and variance = λ.
P(X = k) = (λ^k \* e^–λ)/k!

---

**10. What is a continuous uniform distribution?**
All values within an interval \[a, b] are equally likely.
PDF = 1/(b - a) for x in \[a, b].
Mean = (a + b)/2; Variance = (b - a)²/12.
Used in simulation and modeling random sampling.
A simple example of a continuous distribution.

---

**11. What are the characteristics of a normal distribution?**
Bell-shaped, symmetric around the mean.
Mean = Median = Mode.
Defined by mean (μ) and standard deviation (σ).
Total area under curve = 1.
Used widely in natural and social sciences.

---

**12. What is the standard normal distribution, and why is it important?**
It's a normal distribution with mean 0 and standard deviation 1.
Z-scores are derived from it.
It standardizes data for comparison.
Used in hypothesis testing and confidence intervals.
Tables exist for quick probability lookup.

---

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
CLT states the sampling distribution of the sample mean approaches normality as sample size increases.
Applies regardless of population distribution.
Enables use of normal distribution in inference.
Justifies many statistical tests.
Makes analysis practical with large samples.

---

**14. How does the Central Limit Theorem relate to the normal distribution?**
CLT ensures that the sample mean becomes normally distributed.
This holds even if the original data isn't normal.
Used to approximate probabilities using the normal model.
Facilitates confidence intervals and hypothesis testing.
Improves with larger sample size (n > 30 ideal).

---

**15. What is the application of Z statistics in hypothesis testing?**
Z-statistics test population mean when σ is known.
Used for large samples (n > 30).
Calculated as Z = (X̄ - μ) / (σ/√n).
Compared against standard normal table.
Determines whether to reject H₀.

---

**16. How do you calculate a Z-score, and what does it represent?**
Z = (X - μ) / σ.
It shows how many standard deviations a value is from the mean.
Used to standardize and compare scores.
Z > 0 is above mean; Z < 0 is below.
Commonly used in outlier detection.

---

**17. What are point estimates and interval estimates in statistics?**
Point estimate gives a single value (e.g., sample mean).
Interval estimate provides a range (e.g., confidence interval).
Point is easy but less reliable.
Interval includes uncertainty measure.
Both are used in inferential statistics.

---

**18. What is the significance of confidence intervals in statistical analysis?**
They estimate population parameters with a range.
Give the likelihood that the true value lies within it.
Expressed with a confidence level (e.g., 95%).
Wider intervals = more uncertainty.
Helps in decision-making under uncertainty.

---

**19. What is the relationship between a Z-score and a confidence interval?**
Z-score defines the margin of error in CIs.
E.g., for 95% CI, Z ≈ 1.96.
CI = X̄ ± Z\*(σ/√n).
Z-score determines how wide the interval is.
Used in normal distribution-based inference.

---

**20. How are Z-scores used to compare different distributions?**
Z-scores standardize values across distributions.
Allow comparison regardless of units or scales.
Helps identify relative positions.
Useful in combining or comparing test scores.
Essential in normal-based hypothesis tests.

---

**21. What are the assumptions for applying the Central Limit Theorem?**
Samples must be independent.
Sample size should be sufficiently large (n > 30 usually).
Finite population variance must exist.
Random sampling is required.
Less needed if population is already normal.

---

**22. What is the concept of expected value in a probability distribution?**
Expected value is the long-term average of repetitions.
For discrete: E(X) = Σx·P(x).
For continuous: E(X) = ∫x·f(x)dx.
Represents the center of the distribution.
Used in risk and decision analysis.

---

**23. How does a probability distribution relate to the expected outcome of a random variable?**
It defines the likelihood of each possible value.
Expected value summarizes it as a weighted average.
More likely outcomes contribute more to the expected value.
Helps in forecasting and planning.
Links probability theory with real-world decisions.
