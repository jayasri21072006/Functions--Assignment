1. What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves making an assumption (hypothesis), testing it using sample data, and determining whether to reject or fail to reject the assumption.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
Null hypothesis (H₀): A statement that there is no effect or no difference; it's the default or status quo.

Alternative hypothesis (H₁ or Ha): A statement that contradicts the null and represents the effect or difference you aim to detect.
Example: H₀: μ = 50 vs. H₁: μ ≠ 50

3. What is the significance level in hypothesis testing, and why is it important?
The significance level (α) is the threshold probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01. It determines how strong the evidence must be to reject H₀.

4. What does a P-value represent in hypothesis testing?
A P-value is the probability of obtaining results as extreme as the observed ones, assuming the null hypothesis is true. It quantifies the strength of evidence against H₀.

5. How do you interpret the P-value in hypothesis testing?
If P ≤ α, reject the null hypothesis (evidence supports H₁).

If P > α, do not reject the null hypothesis (not enough evidence).
A small P-value indicates strong evidence against H₀.

6. What are Type 1 and Type 2 errors in hypothesis testing?
Type I Error (α): Rejecting H₀ when it is actually true.

Type II Error (β): Failing to reject H₀ when H₁ is true.
These errors represent false positives and false negatives, respectively.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
One-tailed test: Tests for a deviation in one direction (e.g., μ > μ₀).

Two-tailed test: Tests for any deviation from the null (e.g., μ ≠ μ₀).
Choose based on the research question.

8. What is the Z-test, and when is it used in hypothesis testing?
A Z-test is used when the population standard deviation is known and the sample size is large (n > 30). It assesses how far the sample mean is from the population mean in standard deviation units.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
Z = (X̄ - μ) / (σ / √n)
Where X̄ is the sample mean, μ is the population mean, σ is the population standard deviation, and n is the sample size.
It shows how many standard deviations the sample mean is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown. It is wider than the normal distribution and accounts for more variability.

11. What is the difference between a Z-test and a T-test?
Z-test: Known σ, large n

T-test: Unknown σ, small n
Both test means, but the T-test is more conservative due to increased uncertainty.

12. What is the T-test, and how is it used in hypothesis testing?
A T-test compares sample means to a known value (one-sample), between two independent groups (two-sample), or paired samples (paired T-test) to determine if the means differ significantly.

13. What is the relationship between Z-test and T-test in hypothesis testing?
Both are parametric tests used to compare means. As sample size increases, the T-distribution approximates the Z-distribution. They differ primarily in assumptions about σ and sample size.

14. What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval (CI) gives a range of values within which the true population parameter is likely to fall, with a specified probability (e.g., 95%). If the CI does not include the null value, H₀ may be rejected.

15. What is the margin of error, and how does it affect the confidence interval?
The margin of error is the maximum expected difference between the true population parameter and a sample estimate. Larger margins create wider CIs, showing less precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes’ Theorem updates the probability of a hypothesis based on new evidence. It’s used in decision-making, medical testing, and machine learning.
Formula:
P(A|B) = [P(B|A) × P(A)] / P(B)

17. What is the Chi-square distribution, and when is it used?
The Chi-square (χ²) distribution is used in tests for categorical data like independence or goodness of fit. It’s right-skewed and depends on degrees of freedom.

18. What is the Chi-square goodness of fit test, and how is it applied?
This test checks whether observed frequencies match expected frequencies in a single categorical variable.
Formula:
χ² = Σ[(Oᵢ - Eᵢ)² / Eᵢ]
Compare the result to a critical value to test fit.

19. What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is used to compare two variances. It is also the basis for ANOVA tests. It is right-skewed and used when analyzing multiple groups or regressions.

20. What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) tests if three or more group means differ significantly.
Assumptions:

Independence of observations

Normality of residuals

Equal variances (homoscedasticity)

21. What are the different types of ANOVA tests?
One-Way ANOVA: One factor, multiple groups

Two-Way ANOVA: Two factors, interaction possible

Repeated Measures ANOVA: Same subjects measured multiple times

22. What is the F-test, and how does it relate to hypothesis testing?
An F-test compares variances between two populations or models. It's used in ANOVA and regression to determine if explanatory variables explain significant variance in the data.

## Question 3: Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(k, p):
  """
  Calculates the probability distribution function (PDF) of a Bernoulli distribution.

  Args:
    k: The outcome (0 or 1).
    p: The probability of success (1).

  Returns:
    The probability of the given outcome.
  """
  if k == 1:
    return p
  elif k == 0:
    return 1 - p
  else:
    return "Invalid outcome. For Bernoulli, k must be 0 or 1."

# Example usage:
probability_of_success = 0.7
outcome1 = 1
outcome0 = 0

pdf_outcome1 = bernoulli_pdf(outcome1, probability_of_success)
pdf_outcome0 = bernoulli_pdf(outcome0, probability_of_success)

print(f"PDF for k={outcome1}: {pdf_outcome1}")
print(f"PDF for k={outcome0}: {pdf_outcome0}")
```

## Question 9: Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_zscores(data):
  """
  Calculates Z-scores for a dataset and plots them.

  Args:
    data: A NumPy array or list of numerical data.
  """
  # Calculate Z-scores
  z_scores = zscore(data)

  # Plot the Z-scores
  plt.figure(figsize=(8, 6))
  plt.scatter(range(len(z_scores)), z_scores, alpha=0.6)
  plt.axhline(0, color='red', linestyle='--', linewidth=0.8, label='Mean (Z=0)')
  plt.xlabel("Data Point Index")
  plt.ylabel("Z-score")
  plt.title("Z-scores of the Dataset")
  plt.legend()
  plt.grid(axis='y', alpha=0.5)
  plt.show()

# Example usage:
dataset = [10, 12, 15, 11, 14, 13, 16, 18, 20, 9]
calculate_and_plot_zscores(dataset)
```

## Question 15: Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_mean = 65
sample_std_dev = 10
sample_size = 50

# Hypothesized population mean (null hypothesis)
population_mean_h0 = 60

# Significance level (alpha)
alpha = 0.05

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean_h0) / (sample_std_dev / np.sqrt(sample_size))

# Find the critical Z-value for a two-tailed test
critical_z_value = norm.ppf(1 - alpha/2)

# Determine if the null hypothesis is rejected
print(f"Sample Mean: {sample_mean}")
print(f"Hypothesized Population Mean (H0): {population_mean_h0}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"Critical Z-value (alpha={alpha}): {critical_z_value:.2f}")

if abs(z_statistic) > critical_z_value:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")

# Alternatively, using the p-value
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
print(f"P-value: {p_value:.4f}")

if p_value <= alpha:
  print("Reject the null hypothesis (P-value approach).")
else:
  print("Fail to reject the null hypothesis (P-value approach).")
```

## Question 16: Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy.stats import t

# Sample data
data = [10, 12, 15, 11, 14, 13, 16, 18, 20, 9]

# Confidence level
confidence_level = 0.95

# Calculate the sample mean and standard deviation
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1) # Use ddof=1 for sample standard deviation

# Calculate the standard error
standard_error = sample_std_dev / np.sqrt(len(data))

# Find the critical t-value
degrees_of_freedom = len(data) - 1
alpha = 1 - confidence_level
critical_t_value = t.ppf(1 - alpha/2, degrees_of_freedom)

# Calculate the margin of error
margin_of_error = critical_t_value * standard_error

# Calculate the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Sample Data: {data}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std_dev:.2f}")
print(f"Confidence Level: {confidence_level}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"Critical t-value: {critical_t_value:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"Confidence Interval: ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f})")

print(f"\nInterpretation: We are {confidence_level*100}% confident that the true population mean lies within the interval ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f}).")
```

### Question 1: Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_mean = 65
sample_std_dev = 10
sample_size = 50

# Known population mean (null hypothesis)
population_mean_h0 = 60

# Significance level (alpha)
alpha = 0.05

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean_h0) / (sample_std_dev / np.sqrt(sample_size))

# Find the critical Z-value for a two-tailed test
critical_z_value = norm.ppf(1 - alpha/2)

# Determine if the null hypothesis is rejected
print(f"Sample Mean: {sample_mean}")
print(f"Known Population Mean (H0): {population_mean_h0}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"Critical Z-value (alpha={alpha}): {critical_z_value:.2f}")

if abs(z_statistic) > critical_z_value:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")

# Alternatively, using the p-value
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
print(f"P-value: {p_value:.4f}")

if p_value <= alpha:
  print("Reject the null hypothesis (P-value approach).")
else:
  print("Fail to reject the null hypothesis (P-value approach).")
```

### Question 4: Perform a two-tailed Z-test using Python and visualize the decision region on a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats.weightstats import ztest

# Sample data
sample_data = np.random.normal(loc=52, scale=10, size=100) # Sample size > 30 for Z-test

# Known population mean (null hypothesis)
population_mean_h0 = 50

# Significance level (alpha)
alpha = 0.05

# Perform two-tailed Z-test
z_statistic, p_value = ztest(sample_data, value=population_mean_h0, alternative='two-sided')

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"Known Population Mean (H0): {population_mean_h0}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Find the critical Z-values for a two-tailed test
critical_z_lower = norm.ppf(alpha/2)
critical_z_upper = norm.ppf(1 - alpha/2)

# Visualize the standard normal distribution and decision regions
x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x, loc=0, scale=1)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'k-', lw=2, label='Standard Normal Distribution')
plt.fill_between(x, pdf, where=(x < critical_z_lower) | (x > critical_z_upper), color='red', alpha=0.5, label='Rejection Region (alpha={})'.format(alpha))
plt.axvline(z_statistic, color='blue', linestyle='--', label=f'Observed Z-statistic ({z_statistic:.2f})')

plt.xlabel("Z-score")
plt.ylabel("Density")
plt.title("Two-tailed Z-test and Decision Region")
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.show()

# Interpretation
print("\nInterpretation:")
if abs(z_statistic) > critical_z_upper: # or p_value <= alpha
  print(f"Since the observed Z-statistic ({z_statistic:.2f}) falls in the rejection region (beyond {critical_z_upper:.2f} or below {critical_z_lower:.2f}), we reject the null hypothesis.")
else:
  print(f"Since the observed Z-statistic ({z_statistic:.2f}) falls outside the rejection region, we fail to reject the null hypothesis.")
```

### Question 5: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu_h0, mu_h1, sigma, alpha, sample_size):
    """
    Calculates and visualizes Type 1 and Type 2 errors for a one-tailed Z-test.

    Args:
        mu_h0: Population mean under the null hypothesis.
        mu_h1: Population mean under the alternative hypothesis.
        sigma: Population standard deviation.
        alpha: Significance level (Type 1 error probability).
        sample_size: Sample size.
    """
    se = sigma / np.sqrt(sample_size)

    # Critical value under H0 (for a right-tailed test)
    critical_value = norm.ppf(1 - alpha, loc=mu_h0, scale=se)

    # Type 1 error (alpha) - area under H0 distribution beyond critical value
    x_h0 = np.linspace(mu_h0 - 4*se, mu_h0 + 4*se, 100)
    pdf_h0 = norm.pdf(x_h0, loc=mu_h0, scale=se)
    type1_error = 1 - norm.cdf(critical_value, loc=mu_h0, scale=se)


    # Type 2 error (beta) - area under H1 distribution below critical value
    x_h1 = np.linspace(mu_h1 - 4*se, mu_h1 + 4*se, 100)
    pdf_h1 = norm.pdf(x_h1, loc=mu_h1, scale=se)
    type2_error = norm.cdf(critical_value, loc=mu_h1, scale=se)


    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x_h0, pdf_h0, label=f'Null Hypothesis (H0: μ = {mu_h0})')
    plt.plot(x_h1, pdf_h1, label=f'Alternative Hypothesis (H1: μ = {mu_h1})')
    plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical Value ({critical_value:.2f})')

    # Shade Type 1 error region
    plt.fill_between(x_h0, pdf_h0, where=(x_h0 > critical_value), color='red', alpha=0.5, label=f'Type I Error (α = {type1_error:.4f})')

    # Shade Type 2 error region
    plt.fill_between(x_h1, pdf_h1, where=(x_h1 < critical_value), color='blue', alpha=0.5, label=f'Type II Error (β = {type2_error:.4f})')


    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.title("Type I and Type II Errors in Hypothesis Testing (Right-tailed test)")
    plt.legend()
    plt.grid(axis='y', alpha=0.5)
    plt.show()

# Example usage:
visualize_errors(mu_h0=50, mu_h1=55, sigma=10, alpha=0.05, sample_size=30)
```

### Question 9: Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy.stats import t

def calculate_confidence_interval(data, confidence_level=0.95):
  """
  Calculates the confidence interval for a sample mean.

  Args:
    data: A NumPy array or list of numerical data.
    confidence_level: The desired confidence level (e.g., 0.95 for 95%).

  Returns:
    A tuple containing the lower and upper bounds of the confidence interval.
  """
  sample_mean = np.mean(data)
  sample_std_dev = np.std(data, ddof=1) # Use ddof=1 for sample standard deviation
  sample_size = len(data)

  # Calculate the standard error
  standard_error = sample_std_dev / np.sqrt(sample_size)

  # Find the critical t-value
  degrees_of_freedom = sample_size - 1
  alpha = 1 - confidence_level
  critical_t_value = t.ppf(1 - alpha/2, degrees_of_freedom)

  # Calculate the margin of error
  margin_of_error = critical_t_value * standard_error

  # Calculate the confidence interval
  confidence_interval_lower = sample_mean - margin_of_error
  confidence_interval_upper = sample_mean + margin_of_error

  return confidence_interval_lower, confidence_interval_upper

# Example usage:
dataset = [10, 12, 15, 11, 14, 13, 16, 18, 20, 9]
conf_level = 0.95
lower_bound, upper_bound = calculate_confidence_interval(dataset, conf_level)

print(f"Sample Data: {dataset}")
print(f"Sample Mean: {np.mean(dataset):.2f}")
print(f"Confidence Level: {conf_level}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

print("\nSignificance of the Confidence Interval:")
print(f"A {conf_level*100}% confidence interval means that if we were to take many random samples from the same population and calculate a confidence interval for each sample, approximately {conf_level*100}% of those intervals would contain the true population mean.")
print(f"It provides a range of plausible values for the true population mean based on the sample data.")
```

### Question 11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
# Illustrating Bayes' Theorem with a simple example (e.g., medical testing)

# P(Disease): Prior probability of having the disease
p_disease = 0.01

# P(Positive | Disease): Probability of testing positive given you have the disease (Sensitivity)
p_positive_given_disease = 0.95

# P(Negative | No Disease): Probability of testing negative given you don't have the disease (Specificity)
p_negative_given_no_disease = 0.90

# P(Positive | No Disease): Probability of testing positive given you don't have the disease (False Positive Rate)
p_positive_given_no_disease = 1 - p_negative_given_no_disease

# P(No Disease): Prior probability of not having the disease
p_no_disease = 1 - p_disease

# --- Calculate P(Positive) using the Law of Total Probability ---
# P(Positive) = P(Positive | Disease) * P(Disease) + P(Positive | No Disease) * P(No Disease)
p_positive = (p_positive_given_disease * p_disease) + (p_positive_given_no_disease * p_no_disease)

# --- Apply Bayes' Theorem to find P(Disease | Positive) ---
# P(Disease | Positive) = [P(Positive | Disease) * P(Disease)] / P(Positive)
p_disease_given_positive = (p_positive_given_disease * p_disease) / p_positive

print(f"Prior Probability of Disease: {p_disease:.4f}")
print(f"Probability of testing Positive given Disease (Sensitivity): {p_positive_given_disease:.4f}")
print(f"Probability of testing Negative given No Disease (Specificity): {p_negative_given_no_disease:.4f}")
print(f"Probability of testing Positive: {p_positive:.4f}")
print(f"Posterior Probability of Disease given Positive Test: {p_disease_given_positive:.4f}")

print("\nExplanation of the Bayesian Inference Process:")
print("1. Start with a Prior Probability (P(Disease)): Your initial belief about the probability of the hypothesis (having the disease) before seeing any new evidence.")
print("2. Gather New Evidence (Positive Test): Observe data or results that are relevant to your hypothesis.")
print("3. Determine the Likelihood (P(Positive | Disease)): The probability of observing the evidence given that the hypothesis is true.")
print("4. Determine the Marginal Likelihood (P(Positive)): The overall probability of observing the evidence, considering all possible hypotheses. This acts as a normalizing constant.")
print("5. Apply Bayes' Theorem: Update your prior probability using the likelihood and marginal likelihood to get the Posterior Probability (P(Disease | Positive)). This is your updated belief about the hypothesis after considering the new evidence.")
print("In this example, even with a positive test, the probability of having the disease is still relatively low due to the low prior probability and the possibility of false positives.")
```

### Question 13: Perform a Chi-square test for independence between two categorical variables in Python

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
  """
  Calculates the expected frequencies for a Chi-square test of independence.

  Args:
    observed_data: A NumPy array representing the observed frequencies in a contingency table.

  Returns:
    A NumPy array representing the expected frequencies.
  """
  row_totals = observed_data.sum(axis=1)
  col_totals = observed_data.sum(axis=0)
  grand_total = observed_data.sum()

  expected_frequencies = np.outer(row_totals, col_totals) / grand_total

  return expected_frequencies

# Example observed data
observed_data = np.array([[30, 20, 10],
                          [15, 25, 30]])

expected_freq = calculate_expected_frequencies(observed_data)

print("Observed Data (Contingency Table):\n", observed_data)
print("\nCalculated Expected Frequencies:\n", expected_freq)

# Verification using scipy (optional)
from scipy.stats import chi2_contingency
chi2_statistic, p_value, degrees_of_freedom, expected_frequencies_scipy = chi2_contingency(observed_data)
print("\nExpected Frequencies (from scipy.stats.chi2_contingency):\n", expected_frequencies_scipy)
```

### Question 14: Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (e.g., counts of outcomes in a categorical variable)
observed_counts = np.array([25, 30, 20, 15, 10])

# Expected frequencies (e.g., based on a theoretical distribution or prior knowledge)
# Let's assume a uniform expected distribution for this example
total_counts = observed_counts.sum()
num_categories = len(observed_counts)
expected_counts = np.full(num_categories, total_counts / num_categories)

print(f"Observed Counts: {observed_counts}")
print(f"Expected Counts (assuming uniform distribution): {expected_counts}")

# Perform the Chi-square goodness-of-fit test
chi2_statistic, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

print(f"\nChi-square Statistic: {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation based on a common significance level (e.g., alpha = 0.05)
alpha = 0.05
print("\nInterpretation:")
if p_value <= alpha:
  print(f"At alpha = {alpha}, we reject the null hypothesis. The observed distribution is significantly different from the expected distribution.")
else:
  print(f"At alpha = {alpha}, we fail to reject the null hypothesis. The observed distribution is not significantly different from the expected distribution.")
```
